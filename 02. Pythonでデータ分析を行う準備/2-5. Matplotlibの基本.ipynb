# Matplotlib
- データ可視化のためのPythonパッケージ
- "Seaborn"と一緒に読み込むことでより見やすい図になる

```python
# 読み込み
#   描画系の機能はほとんどpyplotを用いる
import matplotlib.pyplot as plt
import seaborn as sns

# Jupyter Notebook上でグラフ表示を行うためのマジックコマンド
%matplotlib inline
```

```python
# 線グラフを描画する
import numpy as np

# データ生成
# 0〜10まで100感覚の数列を作成
x = np.linspace(0, 10, 100)
# それにランダム（正規分布に従う≒特定の値に偏りやすい）な値を足していく
y = x + np.random.randn(100) 

# プロット
plt.plot(x, y)
```

```python
# 散布図
plt.scatter(x, y)
```

```python
# ヒストグラム
#  データの区間を区切ってそれぞれを集計した値を可視化したグラフ
plt.hist(np.random.randn(1000), bins=10)
```

```python
# 色を分けて描画する

# データの準備
target = np.array([0 for i in range(0,50)] + [1 for i in range(0,50)])
target
```

```python
# 描画
plt.scatter(x, y, c=target)
```
